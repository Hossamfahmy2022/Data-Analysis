```python
import pandas as pd
import numpy as np
import missingno
```

```python
df = pd.read_csv('avocado.csv', index_col=0)
df.isna().sum()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
```

```python
df.info()
```

```python
df.describe()
```

```python
min_bags = df['Total Bags'].min()

df.loc[df['Total Bags']==min_bags, 'Total Bags'] = np.nan
```

```python
df['Total Bags'].isna().sum()
```

```python
df.dtypes
```

```python
## sort values by date
df = df.sort_values('Date', ascending=False)
df
```

```python
## unique value for all colums
categ_cols = df.select_dtypes(include='object').columns.tolist()
categ_cols
```

```python
## unique value for all colums
for col in categ_cols:
    print(f'{col}')
    print(df[col].unique())
    print('*****'*10)
```

```python
df.nunique()
```

```python
df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_')
df.columns
```

```python
df = df.rename(columns={'averageprice': 'price'})
```

```python
## What is the average price for each region for year 2018

df_2018 = df[df['year']==2018]
df_2018.groupby('region')[['price']].mean().sort_values(by='price', ascending=False)
```

```python
df[(df['region']=='Denver')  & (df['year']==2015)]['price'].mean()
```

```python
pd.pivot_table(data=df, columns='region', index='year', values='price', aggfunc='mean')
```

```python
## how much was the total of price for each year for the organic type
df[df['type']=='organic'].groupby('year')['price'].sum()
```

```python
## how much was the total price for each month of each region
```

```python
df_month_region = df.groupby(['month', 'region'])[['price']].sum()
```

```python
df[(df['month']=='February') & (df['year']==2015)]['price'].sum()
```

```python
# Q16: creat a pivot table showing total of price for each month indexed by year and type.

pd.pivot_table(data=df, columns='month', index=['year', 'type'], values='price', aggfunc='sum')
```
