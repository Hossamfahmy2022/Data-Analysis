```python
import numpy as np
import pandas as pd
import missingno
```

```python
df = pd.read_csv('imdb_movies.csv')
df.head()
```

```python
df['Year']= pd.to_numeric(df['Year'], errors='coerce')
df['Year'] = df['Year'].apply(abs)
```

```python
df
```

```python
df.info()
```

```python
## Drop Unneccessaryy cols
## ['ranking of movie', 'certificate', 'DETAIL ABOUT MOVIE','ACTOR 3', 'ACTOR 4']

df_new = df.drop(columns=['ranking of movie', 'certificate', 'DETAIL ABOUT MOVIE','ACTOR 3', 'ACTOR 4'], axis=1)
```

```python
df_new
```

```python
df_new.columns
```

```python
df_new.columns = df_new.columns.str.lower().str.strip()
```

```python
df_new.columns = df_new.columns.str.replace(' ', '_')
```

```python
df_new
```

```python
## what is max & min ratings
df = df_new.copy(deep=True)
```

```python
max_rate = df['rating'].max()
```

```python
df[df['rating']==max_rate]
```

```python
## movies with rating > 9

df[df['rating'] > 9][['movie_name', 'rating']]
```

```python
## top 10 movies per metascore and user rarting
```

```python
df.sort_values(by=['metascore', 'rating'], ascending=False)['movie_name'].head(10)
```
